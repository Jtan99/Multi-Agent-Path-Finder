```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# folder = 'more_walls/low_density/'
folder = 'more_walls/'
# folder = 'more_agents/low_density/'
# folder = 'more_agents/med_density/'
# folder = 'more_agents/high_densitsy/'

dataframe = pd.read_csv(folder+'total_reg.csv', header=None, names=['map_cols', 'map_rows', 'agents', 'nodes_generated', 'nodes_expanded', 'density', 'runtime'])
df = dataframe[['nodes_expanded', 'runtime', 'density']]

dataframe2 = pd.read_csv(folder+'total_cg.csv', header=None, names=['map_cols', 'map_rows', 'agents', 'nodes_generated', 'nodes_expanded', 'density', 'runtime'])
df2 = dataframe2[['nodes_expanded', 'runtime', 'density']]

dataframe3 = pd.read_csv(folder+'total_dg.csv', header=None, names=['map_cols', 'map_rows', 'agents', 'nodes_generated', 'nodes_expanded', 'density', 'runtime'])
df3 = dataframe3[['nodes_expanded', 'runtime', 'density']]

dataframe4 = pd.read_csv(folder+'total_wdg.csv', header=None, names=['map_cols', 'map_rows', 'agents', 'nodes_generated', 'nodes_expanded', 'density', 'runtime'])
df4 = dataframe4[['nodes_expanded', 'runtime', 'density']]
```

```python
m, b = np.polyfit(df['density'], df['runtime'], 1)
m2, b2 = np.polyfit(df2['density'], df2['runtime'], 1)
m3, b3 = np.polyfit(df3['density'], df3['runtime'], 1)
m4, b4 = np.polyfit(df4['density'], df4['runtime'], 1)

ax = plt.gca()
# df.plot(kind='scatter', label='regular cbs', x='density', y="runtime", ax=ax)
plt.plot(df['density'], m*df['density'] + b, label='CBS')
# df2.plot(kind='scatter', label='cg cbs', x='density', y="runtime", color='red', ax=ax)
plt.plot(df2['density'], m2*df2['density'] + b2, color='red', label='CBSH with CG')
# df3.plot(kind='scatter', label='dg cbs', x='density', y="runtime", color='green', ax=ax)
plt.plot(df3['density'], m3*df3['density'] + b3, color="green", label='CBSH with DG')
# df4.plot(kind='scatter', label='wdg cbs', x='density', y="runtime", color='purple', ax=ax)
plt.plot(df4['density'], m4*df4['density'] + b4, color='purple', label='CBSH with WDG')
ax.set_title('Density (increasing obstacles) vs Runtime')
plt.legend()
plt.xlabel('Agent-Density')
plt.ylabel('Runtime')
# plt.ylim()
plt.show()
```
